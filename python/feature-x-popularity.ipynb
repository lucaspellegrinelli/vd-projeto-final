```python
import pandas as pd
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import re

other_views = pd.read_csv("other_kpop_songs.csv")
top_views = pd.read_csv("top_kpop_songs.csv")

sp = spotipy.Spotify(client_credentials_manager=SpotifyClientCredentials(
    "a5a06f90f9a04b259024dfdeebd837b6",
    "8143816647f94d36aba4bd1fceb203be"
))

other_views.head()
```

```python
song_stats = []

for views_df in [other_views, top_views]:
    for i in range(len(views_df)):
        song_id = views_df.iloc[i].id
        song_features = sp.audio_features([song_id])[0]

        song_stats.append({
            'acousticness': song_features["acousticness"],
            'danceability': song_features["danceability"],
            'energy': song_features["energy"],
            'liveness': song_features["liveness"],
            'speechiness': song_features["speechiness"],
            'valence': song_features["valence"],
            "views": views_df.iloc[i].views
        })
        
song_stats
```

```python
import json
with open("song-stats.json", "w") as file:
    json.dump(song_stats, file)
```
